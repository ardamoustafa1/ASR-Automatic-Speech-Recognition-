# ASR-Pro Python SDK & REST API Demo

This notebook demonstrates how to interact with the ASR-Pro backend using pure Python and `requests`. We will authenticate, upload an audio file for batch processing, and query the analytics engine.

```python
import json

import requests

BASE_URL = "http://localhost:8000/api/v1"
USERNAME = "admin"
PASSWORD = "admin" # Default if not changed via env

# 1. Authenticate
response = requests.post(f"{BASE_URL}/auth/login", data={"username": USERNAME, "password": PASSWORD})
response.raise_for_status()
token = response.json()["access_token"]
headers = {"Authorization": f"Bearer {token}"}
print(f"Successfully authenticated! Token: {token[:10]}...")
```

## 2. Analyze a text transcript
In a real scenario, you'd send audio via WebSocket or upload a file. Here we send a raw transcript for Zero-Shot intelligence processing.

```python
payload = {
    "external_id": "CRM-999123",
    "agent_id": "AGENT-007",
    "customer_id": "CUST-100",
    "sector": "omni",
    "duration_sec": 45.0,
    "full_transcript": "Alo merhabalar faturam çok yüksek gelmiş itiraz etmek istiyorum iptal edin bu üyeliği aksi takdirde sizi mahkemeye veririm."
}

res = requests.post(f"{BASE_URL}/conversations/analyze", json=payload, headers=headers)
print(json.dumps(res.json(), indent=2, ensure_ascii=False))
```

## 3. Query Real-Time Trends
Fetch current keyword anomalies and dashboard statistics.

```python
dashboard_res = requests.get(f"{BASE_URL}/analytics/dashboard", headers=headers)
print("Dashboard Stats:", dashboard_res.json())

top_keywords_res = requests.get(f"{BASE_URL}/analytics/top-keywords?limit=5", headers=headers)
print("\nTop Keywords:", json.dumps(top_keywords_res.json(), indent=2, ensure_ascii=False))
```
